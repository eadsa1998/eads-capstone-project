```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
#Importing Dataset
df = pd.read_csv("dream_text_corpus.csv", index_col=0)
#Checking the Data
df.head()
#Removing NaN Values
#Creating the text variable
text = " ".join(dream for dream in df.dreams)
# Get stopwords
stopwords = set(STOPWORDS)
stopwords.update(['back','go','one','said','going','went','see','got','something','u','say','thing','around','now','want'])
# Creating word_cloud with text as argument in .generate() method
word_cloud = WordCloud(stopwords=stopwords, collocations = False, background_color = 'black').generate(text)
# Display the generated Word Cloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
```

```python
word_cloud.to_file("wordcloud.png")
```
